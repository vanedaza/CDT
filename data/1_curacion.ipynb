```python
id_vvv = vivas.VIVAID.values
display(len(id_vvv))

for target in id_vvv:

    ruta_d078 = 'd078_lcs_TM/'+ str(int(target)) +'.dat'
    ruta_TARGET = 'TimeSeries_vivace/'+ str(int(target))
    shutil.copy(ruta_d078, ruta_TARGET)
    
print(len(listdir('TimeSeries_vivace/')))
```

```python
columns=['VIVAID', 'ra(J2000)', 'dec(J2000)','ksEMeanMagPawprint',
         'FreqKfi2', 'FreqLfl2', 'FreqLSG', 'FreqPDM', 'FreqSTR']
df3 = vivas[columns]
df3.head(5)
```

```python
df3['Period_KFI'] = df3['FreqKfi2'].apply(lambda x: 1/x)
df3['Period_PAN'] = df3['FreqLfl2'].apply(lambda x: 1/x)
df3['Period_LSG'] = df3['FreqLSG'].apply(lambda x: 1/x)
df3['Period_PDM'] = df3['FreqPDM'].apply(lambda x: 1/x)
df3['Period_STR'] = df3['FreqSTR'].apply(lambda x: 1/x)
```
